```python
%matplotlib inline
import numpy as np
import itertools
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.io import savemat

BLUE = '#0074D9'
RED = '#FF4136'
```

## №5. Непрерывная функция одной переменной

```python
def fun(x):
    return (np.sin(50 * x) + np.sin(25 * x) + np.sin(10 * x) + np.cos(10 * x) + \
        np.exp(x - 1) + np.cosh(3 * (1-x)) + np.cosh(3 * x)) / 15

X = np.linspace(0, 1, 500)
Y = fun(X)
```

```python
plt.figure(figsize=(6, 2), dpi=120)
plt.plot(X, Y, '-', c=BLUE)
plt.xlabel('$x$')
plt.ylabel('$y=f(x)$')
plt.savefig('../pics/1_1', bbox_inches='tight')
plt.show()
```

```python
N = 100
P = np.linspace(0, 1, N)
T = fun(P)
savemat('matlab/5.mat', mdict={'P': P, 'T': T})
```
